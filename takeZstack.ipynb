```python
import flashenscope as fs
mmc = fs.initialize()

optovar = '1x'

# from flashenscope.core import reload_core
# reload_core()
# print('core reload done')
```

```python
# live view
config = 'dapi_10x'
fs.live(config)
```

```python
import os
file_dir = 'pollen_Henry/2025-12-11/'
# filename_noExt = '685_1h_styleStart'
# filename_noExt = '685_1h_pGrainCounts'
filename_noExt = '687_6h_styleMid'
filename = file_dir + filename_noExt + '.tif'
startPosZ = 200
endPosZ = 215
z_inc = 20
```

```python
# take imgs
import time
import numpy as np
import tifffile
import json

if os.path.isfile(filename):
    print("File exists, rename")
else:
    zPosArray = np.arange(startPosZ, endPosZ + z_inc, z_inc)
    print(zPosArray.size)

    for i, zpos in enumerate(zPosArray, start=1):

        mmc.setZPosition(np.double(zpos))
        mmc.waitForSystem()
        time.sleep(0.05)
    
        img = fs.takePicture(config)   # returns a numpy array
    
        # Append to multipage TIFF
        tifffile.imwrite(filename, img, append=True)
    
        print(f"{i} out of {len(zPosArray)} images written")

    # save metadata
    
    meta = {'Config': config,'config_metadata': mmc.getTags(),
        'optovar': optovar, 'pixel_size_um': mmc.getPixelSizeUm(),
        'Timestamp': time.strftime('%Y-%m-%d %H:%M:%S'),
        'ZStack_spacing_um': z_inc, 'startPosZ': startPosZ, 'endPosZ': endPosZ,
        'zPosArray': zPosArray.tolist(), 'filename': filename_noExt, 'dir': file_dir,
    }
    with open(file_dir+filename_noExt+'_metadata.json', 'w') as mf:
        json.dump(meta, mf, indent=2)
    print(filename_noExt+" done")
```
